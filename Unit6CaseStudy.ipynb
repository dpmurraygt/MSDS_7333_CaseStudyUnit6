### Feature definitions for cleaning:

t = timestamp (timestamp in ms since Unix epoch | posix time)

id = MACofScanDevice (cell phone in this case)

pos = RealPosition (x,y physical coordinate)

degree = orientation (of scanning device)

MACofResponse1 .. N = Signal Strength Value, Frequency, Mode  (responding peer MAC of access point or device in adhoc mode for example) where signal strength is in dBm, channel frequency and its mode (AP = 3, adhoc = 1) 

we are only interested in APs

```r
# return null if no signals
processLine = function(x)
    {
    tokens = strsplit(x, "[;=,]")[[1]]
    if(length(tokens) == 10)
        return(NULL)
    handheld = tokens[c(2,4,6:8,10)]
    readings = matrix(tokens[-(1:10)], ncol=4, byrow=TRUE)
    handheld_matrix = matrix(handheld, nrow=nrow(readings), ncol=length(handheld), byrow=TRUE)
    cbind(handheld_matrix, readings)
}
```

```r
options(error=recover, warn=1)
tmp = lapply(lines, processLine)
offline = as.data.frame(do.call("rbind", tmp),
                       stringsAsFactors=FALSE)
```
